### MEMBACA GAMBAR DAN INISIALISASI WARNA

```python
import cv2 as cv #202431129_Gian Ananda Virja
import numpy as np
import matplotlib.pyplot as plt

gambar_tulisan = cv.imread('tulisan_gian.jpeg')
b = gambar_tulisan[:, :, 0]
g = gambar_tulisan[:, :, 1]
r = gambar_tulisan[:, :, 2]

panjang = len(gambar_tulisan)
lebar = len(gambar_tulisan[0])
```

### NILAI AMBANG NONE

```python
ambang = 255 #202431129_Gian Ananda Virja
gambar_none = np.zeros((panjang, lebar), dtype=np.uint8)

for i in range(panjang):
    for j in range(lebar):
        nilai_gray = int(0.299 * r[i,j] + 0.587 * g[i,j] + 0.114 * b[i,j])
        if nilai_gray > ambang:
            gambar_none[i, j] = 1
        else:
            gambar_none[i, j] = 0
```

### NILAI AMBANG BIRU

```python
ambang_bawah_biru = 95 #202431129_Gian Ananda Virja
ambang_atas_biru = 255
gambar_b = np.zeros((panjang, lebar), dtype=np.uint8)

for i in range(panjang):
    for j in range(lebar):
        if ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and g[i, j] < 100 and r[i, j] < 100:
            gambar_b[i, j] = 1
        else:
            gambar_b[i, j] = 0
```

### NILAI AMBANG BIRU DAN MERAH

```python
ambang_bawah_merah = 120 #202431129_Gian Ananda Virja
ambang_atas_merah = 255
gambar_rb = np.zeros((panjang, lebar), dtype=np.uint8)

for i in range(panjang):
    for j in range(lebar):
        biru_flag = ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and r[i, j] < 100 and g[i, j] < 100
        merah_flag = ambang_bawah_merah <= r[i, j] <= ambang_atas_merah and g[i, j] < 100 and b[i, j] < 100
        if merah_flag or biru_flag:
            gambar_rb[i, j] = 1
        else:
            gambar_rb[i, j] = 0
```

### NIlAI AMBANG BIRU, HIJAU DAN MERAH SERTA HASIL CITRA

```python
ambang_bawah_hijau = 95 #202431129_Gian Ananda Virja
ambang_atas_hijau = 255
gambar_rgb = np.zeros((panjang, lebar), dtype=np.uint8)

for i in range(panjang):
    for j in range(lebar):
        biru_flag = ambang_bawah_biru <= b[i, j] <= ambang_atas_biru and r[i, j] < 100 and g[i, j] < 100
        merah_flag = ambang_bawah_merah <= r[i, j] <= ambang_atas_merah and g[i, j] < 100 and b[i, j] < 100
        hijau_flag = ambang_bawah_hijau <= g[i, j] <= ambang_atas_hijau and b[i, j] < 100 and r[i, j] < 100
        if biru_flag or merah_flag or hijau_flag:
            gambar_rgb[i, j] = 1
        else:
            gambar_rgb[i, j] = 0

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].imshow(gambar_none, cmap='gray')
axs[0, 0].set_title("NONE")
axs[0, 0].axis('off')

axs[0, 1].imshow(gambar_b, cmap='gray')
axs[0, 1].set_title("BLUE")
axs[0, 1].axis('off')

axs[1, 0].imshow(gambar_rb, cmap='gray')
axs[1, 0].set_title("RED-BLUE")
axs[1, 0].axis('off')

axs[1, 1].imshow(gambar_rgb, cmap='gray')
axs[1, 1].set_title("RED-GREEN-BLUE")
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

## MEMPERBAIKI backlight (SOAL2)

### IMPORT LIBRARY DAN MEMBACA GAMBAR

```python
import cv2 as cv #202431129_Gian Ananda Virja
import numpy as np
import matplotlib.pyplot as plt

gambar_asli = cv.imread('selfie_gian.jpeg')
r, g, b = gambar_asli[:,:,2], gambar_asli[:,:,1], gambar_asli[:,:,0]
panjang = len(gambar_asli)
lebar = len(gambar_asli[0])
```

### KONVERSI GAMBAR KE GRAYSCALE

```python
gambar_gray = np.zeros((panjang, lebar), dtype=np.uint8) #202431129_Gian Ananda Virja
for i in range(panjang):
    for j in range(lebar):
        gambar_gray[i,j] = int(0.299 * r[i,j] + 0.587 * g[i,j] + 0.114 * b[i,j])
```

### MANIPULASI KECERAHAN DENGAN GAMMA

```python
gambar_cerah = np.zeros((panjang, lebar), dtype=np.uint8) #202431129_Gian Ananda Virja
gamma = 1.4  
for i in range(panjang):
    for j in range(lebar):
        hasil = ((gambar_gray[i,j] / 255.0) ** (1.0 / gamma)) * 255
        gambar_cerah[i,j] = np.clip(hasil, 0, 255)
```

### PEREGANGAN KONTRAS

```python
gambar_kontras = np.zeros((panjang, lebar), dtype=np.uint8) #202431129_Gian Ananda Virja
kontras = 1.25  
for i in range(panjang):
    for j in range(lebar):
        hasil = 128 + (int(gambar_gray[i,j]) - 128) * kontras
        gambar_kontras[i,j] = np.clip(hasil, 0, 255)
```

### KOMBINASI PENCERAHAN DAN KONTRAS

```python
gambar_final = np.zeros((panjang, lebar), dtype=np.uint8) #202431129_Gian Ananda Virja
gamma_f = 1.25
kontras_f = 1.15
for i in range(panjang):
    for j in range(lebar):
        cerah_f = ((gambar_gray[i,j] / 255.0) ** (1.0 / gamma_f)) * 255
        hasil = 128 + (cerah_f - 128) * kontras_f
        gambar_final[i,j] = np.clip(hasil, 0, 255)

fig, axs = plt.subplots(3, 2, figsize=(11, 14))
axs[0,0].imshow(cv.cvtColor(gambar_asli, cv.COLOR_BGR2RGB))
axs[0,0].set_title("Gambar Asli")
axs[0,1].imshow(gambar_gray, cmap='gray')
axs[0,1].set_title("Gambar Gray")
axs[1,0].imshow(gambar_cerah, cmap='gray')
axs[1,0].set_title("Gambar Gray yang Dipercerah (Gamma)")
axs[1,1].imshow(gambar_kontras, cmap='gray')
axs[1,1].set_title("Gambar Gray yang Diperkontras (Midpoint)")
axs[2,0].imshow(gambar_final, cmap='gray')
axs[2,0].set_title("Gambar Gray yang Dipercerah dan Diperkontras")
axs[2,1].axis('off')
plt.tight_layout()
plt.show()
```
